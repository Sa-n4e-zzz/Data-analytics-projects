### Шаг 1. Обзор данных

```python
import pandas as pd #импортируем библиотеку
data = pd.read_csv('/datasets/data.csv') #читаем файл и сохраняем в переменную
data.info()#получаем информации о файле
display(data.head(50))
display(data.tail(50))
#просматриваем первые и последние строчки таблицы
#просматриваем первые и последние строчки таблицы
```

### Шаг 2.1 Заполнение пропусков

```python
display(data.isna().sum())
data['total_income'] = data['total_income'].fillna(data['total_income'].median())# заполнили пропуски нужного столбца.
data.isna().sum() # проверили, устранили пропуски.
 
```

1)Пропущенные значения имеются сразу в двух столбцах и их одинаковое количество.

2)Пропущенные значения в каждом из столбцов с пропусками составляют  примерно 10% от всей таблицы.

3)Заполнить пропуски медианным значением — лучшее решение для количественных переменных, потому что: 1) нам необохдимо заполнить пропуски для анализа 2) Cреднее значение не явялется лучшим решением, так как медиана дает более реалистичную оценку центрального значения.

4)Причина отсутвия 10% данных столбца "income_type" связана с недостующими данными в столбце "days_employed", также мы можем увидеть, что в "days_employed" некотрве данные имеют невомозжные значения. 

### Шаг 2.2 Проверка данных на аномалии и исправления.

```python
data['days_employed'] = data['days_employed'].abs() #изменили отрицательные значения функией abs()
data['days_employed'] = data['days_employed'].fillna(data['days_employed'].median())# заполняем пропуски медианными значениями
#display(data['dob_years'].value_counts()) # возраст  некоторых клиентов равен 0, что ошибка, заменем медианным значением.
data.loc[data['dob_years'] == 0,'dob_years'] = data['dob_years'].median()
data['dob_years'] = data['dob_years'].astype('int')
#display(data['dob_years'].value_counts())# проверили,изменилось.
data['gender'].value_counts()# выявилась одна аномалия, однако она нам не помешает при анализе, не будем ее изменять.
data['children'].value_counts()# появилось два анамальных значения, вероятнее это ошибка заполнявшего, попробуем устранить
data.loc[data['children'] == 20,'children'] = 2
data.loc[data['children'] == -1,'children'] = 1
display(data.isna().sum())
data.head(50) # проверяем изменения
```

Причинами  появления таких данных могут появится из-за человеческих факторы, либо в результате технических ошибок(например перенос данных). 
Также, помимо отрицательных значений и пропусков, в некоторых строчках появляются значения, которые невозмможны. Однако эти данные никак не повлияют на наш анализ, поэтому мы не будем трогать их.


### Шаг 2.3. Изменение типов данных.

```python
data['total_income'] = data['total_income'].astype('int') 
display(data)
data.info()
# заменили тип данных на целочисленный и проверили изменения
```

### Шаг 2.4. Удаление дубликатов.

```python
display(data.duplicated().sum())
display(data['education'].value_counts())# необходимо обработать
#display(data['income_type'].value_counts())# в этом столбце все в порядке
#display(data['purpose'].value_counts())# много значений, их мы обработаем чуть позже
#data['family_status'].value_counts()
data['family_status'] = data['family_status'].str.lower()
data['education'] = data['education'].str.lower()# привел к нижнему регистру
display(data['education'].value_counts())#исправили
data = data.drop_duplicates().reset_index(drop=True)
display(data.duplicated().sum())
data.head(50)

```

1) Для удаления дубликатов мы воспользовались: метод drop_duplicates для явных дубликатов, затем value_counts для уникальных значений и в после этого пришлось ручнымм метод переводить строки к нижнему регистру. Таким образом, мы избавились от всех дубликатов и улучшили таблицу
2) Главной причиной появления дубликатов в данной ситуации - человеческий фактор, некоторые значения были написаны по-разному, хоть и имели один и тот же смысл.

### Шаг 2.6. Категоризация дохода.

```python
def total_income_category(row):
    total_income = row['total_income']
    if total_income <= 30000:
        return 'E'
    if total_income <= 50000:
        return 'D'
    if total_income <= 200000:
        return 'C'
    if total_income <= 1000000:
        return 'B'
    return 'A'
data['total_income_category'] = data.apply(total_income_category, axis=1)
display(data['total_income_category'].value_counts())
data
# создали функцию для распределения на категории в зависимости от уровня дохода и добавили новый столбец
```

### Шаг 2.7. Категоризация целей кредита.

```python
data['purpose'].value_counts().sum()#проверяем количество уникальных значений в столбце purpose
def purpose_category(purpose):
    if 'автомобил' in purpose:
        return 'операции с автомобилем'
    if 'недвижимост'in purpose or 'жиль' in purpose:
        return 'операции с недвижимостью'
    if 'свадьб' in purpose:
        return 'проведение свадьбы'
    if 'образов' in purpose:
        return 'получение образования'
data['purpose_category'] = data['purpose'].apply(purpose_category)
#создали функцию, которая категоризирует цели кредита.
display(data['purpose_category'].value_counts())
display(data['purpose_category'].value_counts().sum())# количество уникальных значений в столбце purpose_category сходится со столбцом purpose
data

    
    
```
